```python
df['GPA'].unique()
```

```python
df['GPA']=df['GPA'].str.replace(r'[^\d\.\d+]','').replace( '',np.nan).astype(float).round(2)
df['GPA']=df['GPA'].fillna(df['GPA'].mean())
```

```python
sns.boxplot(df['GPA'])
```

```python
df_new=df[df['GPA']>=2.50]
```

```python
sns.boxplot(df_new['GPA'])
```

```python
food_cat=['ethnic_food','greek_food','indian_food','italian_food','persian_food',
'thai_food']
c=sns.diverging_palette(240, 10, n=9)
sns.heatmap(df_new[food_cat].corr(method='spearman'),cmap=c, annot=True)
```

Obtaining correlation between diet type and GPA

```python
diet_type=pd.get_dummies(df_new['diet_current_coded'])
diet_type=diet_type.rename(columns={1:'healthy/balanced/moderated/', 2:'unhealthy/cheap/too much/random/',3:'the same thing over and over',4:'unclear'})
diet_type['GPA']=df_new['GPA']
c=sns.diverging_palette(50, 10, as_cmap=True)
sns.heatmap(diet_type.corr(method='spearman'),cmap=c, annot=True)
```

```python
ax=sns.countplot(df_new['Gender'])
ax.set_xticklabels(['Female','Male'],rotation=45,fontsize=10)
```

```python
gender_type=pd.get_dummies(df_new['Gender'])
gender_type=gender_type.rename(columns={1:'Female', 2:'Male'})
gender_type['GPA']=df_new['GPA']
c=sns.diverging_palette(50, 10, as_cmap=True)
sns.heatmap(gender_type.corr(method='spearman'),cmap=c, annot=True)
```
